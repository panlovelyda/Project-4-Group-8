```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_regression
```

```python
data_df=pd.read_csv(Path('Shill Bidding Dataset.csv'))
data_df.head()
```

```python
train_df = pd.read_csv(Path('Shill Bidding Dataset.csv'))
test_df = pd.read_csv(Path('Shill Bidding Dataset.csv'))
y_train = train_df["Class"]
X_train = train_df.drop(columns = ["Class"])
X_train = pd.get_dummies(X_train)
X_train.head()
```

```python
y_test = test_df["Class"]
X_test = test_df.drop(columns = ["Class"])
X_test = pd.get_dummies(X_test)
X_test.head()
```

```python
for column in X_train.columns:
    if column not in X_test.columns:
        X_test[column] = 0
reg = LogisticRegression().fit(X_train, y_train)
print("Logistic Regression model score: ", reg.score(X_test, y_test))
```

```python
randomForestClass = RandomForestClassifier(random_state=0)
randomForestClass.fit(X_train, y_train)
print("Random Forest Classifier score: ", randomForestClass.score(X_test,y_test))
```
